```python
#to organize and manipulate the data pandas are used
#numpy performs mathematical and logical operations on multidimensional arrays and metrices
# importing libraries
import pandas as pd
import numpy as np
```

```python
# reads the data from csv file
data = pd.read_csv("austin_weather.csv",parse_dates=["Date"])
data.info()
```

```python
# some values have 'T' which denotes trace rainfall and '-' which denotes no or NIL
# we need to replace all occurrences of T and - with 0 
data=data.drop(['Events','Date'],axis=1)
data=data.replace('T',0)
data=data.replace('-',0)
```

```python
# save the data in a csv file 
data.to_csv('austin_final.csv')
```

```python
#converting features with respect to datatype
data.DewPointHighF=data.DewPointHighF.astype(int)
data.DewPointAvgF=data.DewPointAvgF.astype(int)
data.DewPointLowF=data.DewPointLowF.astype(int)
data.HumidityHighPercent=data.HumidityHighPercent.astype(int)
data.HumidityAvgPercent=data.HumidityAvgPercent.astype(int)
data.HumidityLowPercent=data.HumidityLowPercent.astype(int)
data.SeaLevelPressureHighInches=data.SeaLevelPressureHighInches.astype(float)
data.SeaLevelPressureAvgInches=data.SeaLevelPressureAvgInches.astype(float)
data.SeaLevelPressureLowInches=data.SeaLevelPressureLowInches.astype(float)
data.VisibilityHighMiles=data.VisibilityHighMiles.astype(int)
data.VisibilityAvgMiles=data.VisibilityAvgMiles.astype(int)
data.VisibilityLowMiles=data.VisibilityLowMiles.astype(int)
data.WindHighMPH=data.WindHighMPH.astype(int)
data.WindAvgMPH=data.WindAvgMPH.astype(int)
data.WindGustMPH=data.WindGustMPH.astype(int)
data.PrecipitationSumInches=data.PrecipitationSumInches.astype(float)
```

```python
#dividing the data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#initialize a linear regression classifier 
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
# train the classifier with input data.
regressor.fit(x_train,y_train)
```

```python
#the smaller the means squared error, the closer you are to finding the line of best fit.
from sklearn.metrics import mean_squared_error
y_predict=regressor.predict(x_test)
accuracy=mean_squared_error(y_test,y_predict)
print(accuracy)
```

```python
#RMSE shows how concentrated the data is around the line of best fit.
from sklearn import metrics
y_pred = regressor.predict(x_test)
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
r_square=regressor.score(x,y)
variance_inflation_factor=1/(1-r_square)
print(variance_inflation_factor)
```
